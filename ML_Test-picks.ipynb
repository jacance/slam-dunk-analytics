```python
from config import user, pw, port
from sqlalchemy import create_engine
from sqlalchemy.ext.automap import automap_base
```

```python
engine = create_engine(f'postgresql://{user}:{pw}@localhost:{port}/slam_dunk_analytics_db')

conn = engine.connect()
```

```python
import pandas as pd
pd.set_option('display.max_columns', None)
draftsql = pd.read_sql_query("SELECT * FROM draft", con=engine)
draftsql
```

```python
careersql = pd.read_sql_query("SELECT * FROM career", con=engine)
careersql
```

```python
combined_df = pd.merge(careersql, draftsql, how='left', left_on='new_key', right_on='official_key')

combined_df
```

```python
combined_df.columns
```

```python
combined_df = combined_df[['School', 'Conf', 'G', 'GS', 'MP',
       'FG', 'FGA', 'FGPCT', '2P', '2PA', '2PPCT', '3P', '3PA', '3PPCT', 'FT',
       'FTA', 'FTPCT', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS',
       'player_x', 'last_season', 'unique_player_key', 'draft_year', 'new_key',
       'round', 'pick',]]
combined_df
```

```python
combined_df = combined_df.fillna(0)
combined_df
```

```python
combined_df['round'] = combined_df['round'].astype('int64')
combined_df['pick'] = combined_df['pick'].astype('int64')

columns = ['G', 'GS', 'MP',
       'FG', 'FGA', 'FGPCT', '2P', '2PA', '2PPCT', '3P', '3PA', '3PPCT', 'FT',
       'FTA', 'FTPCT', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']
```

```python
import numpy as np
# X = combined_df[columns]
# y = combined_df['pick'].values.reshape(-1, 1)
# print(X.shape, y.shape)
```

```python
X = combined_df[columns].loc[(combined_df['draft_year'] > 2004) & (combined_df['draft_year'] < 2021)]
X2021 = combined_df[columns].loc[combined_df['draft_year'] == 2021]
y = combined_df['pick'].loc[(combined_df['draft_year'] > 2004) & (combined_df['draft_year'] < 2021)].values.reshape(-1,1)
y2021 = combined_df['round'].loc[combined_df['draft_year'] == 2021]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

### Lasso

```python
from sklearn import linear_model
reg = linear_model.Lasso(alpha=0.1)
reg.fit(X_train_scaled, y_train)
```

```python
reg.predict(X_test_scaled)
```

## Decision Tree

```python
from sklearn import tree

clf = tree.DecisionTreeClassifier(class_weight = {})
clf = clf.fit(X_train_scaled, y_train)
print("DecisionTreeClassifier Train score", clf.score(X_train_scaled, y_train))
print("DecisionTreeClassifier Test score", clf.score(X_test_scaled, y_test))
```

```python
tr_predictions = clf.predict(X_train_scaled)
te_predictions = clf.predict(X2021)
print(f"Train: {tr_predictions}")
print(f"Test: {te_predictions}")
```

```python
np.unique(te_predictions, return_counts = True)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf = rf.fit(X_train_scaled, y_train)
print("RandomForestClassifier Train score", rf.score(X_train_scaled, y_train))
print("RandomForestClassifier Test score", rf.score(X_test_scaled, y_test))
```

```python
train_predictions = rf.predict(X_train_scaled)
test_predictions = rf.predict(X2021)
print(f"Train: {train_predictions}")
print(f"Test: {test_predictions}")
```

```python
import numpy as np
np.unique(test_predictions, return_counts = True)
```

```python
X_train_indexes = list(X_train.index)
X_test_indexes = list(X_test.index)
train_predictions1 = list(train_predictions)
test_predictions1 = list(test_predictions)
predicted_round_list = []

train_counter = 0
test_counter = 0
for i in range(len(combined_df)):
    if i in X_train_indexes:
        predicted_round_list.append(train_predictions1[X_train_indexes.index(i)])
    if i in X_test_indexes:
        predicted_round_list.append(test_predictions1[X_test_indexes.index(i)])
```

```python
combined_df['predicted_pick'] = predicted_round_list
```

```python
combined_df[combined_df['predicted_pick'] == 1 & (combined_df['draft_year'] == 2021)]
```

## Neural Network

```python
from tensorflow.keras.models import Sequential

model2 = Sequential()
```

```python
from tensorflow.keras.layers import Dense
number_inputs = 22
number_hidden_nodes = 8
model2.add(Dense(units=number_hidden_nodes,
                activation='relu', input_dim=number_inputs))

model2.add(Dense(units=number_hidden_nodes,
                activation='relu'))

number_classes = 3
model2.add(Dense(units=number_classes, activation='softmax'))
```

```python
model2.summary()
```

```python
model2.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model2.fit(
    X_train_scaled,
    y_train,
    epochs=50,
    shuffle=True,
    verbose=2
)
```

```python
model2_loss, model2_accuracy = model2.evaluate(
    X_test_scaled, y_test, verbose=2)
print(f"Normal Neural Network - Loss: {model2_loss}, Accuracy:{model2_accuracy}")
```

```python
train_predictions = model2.predict_classes(X_train_scaled)
test_predictions = model2.predict_classes(X_test_scaled)
print(f"Train: {train_predictions}")
print(f"Test: {test_predictions}")
```
