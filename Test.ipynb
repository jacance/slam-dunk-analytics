```python
#letters = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']

#for i in letters:
url = 'https://www.sports-reference.com/cbb/players/z-index.html'
response = requests.get(url)
soup = bs(response.text, 'html.parser')
href = soup.find_all('a')

all_links = [x['href'] for x in href]

player_links = []

for i in all_links:
    try:
        if '/cbb/players' in i:
            player_links.append(i)
    except:
        pass

individual_player_links = [x.replace("/cbb/players","") for x in player_links if x != '/cbb/players']
final_player_links = [x for x in individual_player_links if x != '/']
final_player_links
```

```python
import bs4
url2 = f'https://www.sports-reference.com/cbb/players{final_player_links[0]}'
response2 = requests.get(url2)
soup2 = bs(response2.text)
comments = soup2.find_all(text=lambda text:isinstance(text, bs4.Comment))
per_40_html = []
for i in range(len(comments)):
    if '<table class="row_summable sortable stats_table" id="players_per_min" data-cols-to-freeze=",1">' in str(comments[i]):
        per_40_html.append(comments[i])
    else:
        pass
table = pd.read_html(per_40_html[0])
table[0]
```
