```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
careersql = pd.read_csv('career.csv')
draftsql = pd.read_csv('1993later.csv')
```

```python
combined_df = pd.merge(careersql, draftsql, how='left', left_on='new_key', right_on='official_key')

combined_df
```

```python
combined_df.columns
```

```python
combined_df = combined_df[['School', 'Conf', 'G', 'GS', 'MP',
       'FG', 'FGA', 'FGPCT', '2P', '2PA', '2PPCT', '3P', '3PA', '3PPCT', 'FT',
       'FTA', 'FTPCT', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS',
       'player_x', 'last_season', 'unique_player_key', 'draft_year', 'new_key',
       'round', 'pick',]]
combined_df
```

```python
combined_df = combined_df.fillna(0)
combined_df
```

```python
combined_df.columns
```

```python
X = combined_df[['G', 'GS', 'MP',
       'FG', 'FGA', 'FGPCT', '2P', '2PA', '2PPCT', '3P', '3PA', '3PPCT', 'FT',
       'FTA', 'FTPCT', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS',]]
y = combined_df['round'].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

## Decision Tree

```python
from sklearn import tree

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train_scaled, y_train)
print("DecisionTreeClassifier Train score", clf.score(X_train_scaled, y_train))
print("DecisionTreeClassifier Test score", clf.score(X_test_scaled, y_test))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train_scaled, y_train)
print("RandomForestClassifier Train score", rf.score(X_train_scaled, y_train))
print("RandomForestClassifier Test score", rf.score(X_test_scaled, y_test))
```

## Neural Network

```python
from tensorflow.keras.models import Sequential

model = Sequential()
```

```python
from tensorflow.keras.layers import Dense
number_inputs = 22
number_hidden_nodes = 4
model.add(Dense(units=number_hidden_nodes,
                activation='relu', input_dim=number_inputs))

number_classes = 3
model.add(Dense(units=number_classes, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(
    X_train_scaled,
    y_train,
    epochs=1000,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
import numpy as np
train_predictions = model.predict_classes(X_train_scaled)
test_predictions = model.predict_classes(X_test_scaled)
print(f"Train: {train_predictions}")
print(f"Test: {test_predictions}")
```

```python
X_train_indexes = list(X_train.index)
X_test_indexes = list(X_test.index)
train_predictions1 = list(train_predictions)
test_predictions1 = list(test_predictions)
predicted_round_list = []

train_counter = 0
test_counter = 0
for i in range(len(combined_df)):
    if i in X_train_indexes:
        predicted_round_list.append(train_predictions1[X_train_indexes.index(i)])
    if i in X_test_indexes:
        predicted_round_list.append(test_predictions1[X_test_indexes.index(i)])
```

```python
combined_df['predicted_round'] = predicted_round_list
```
