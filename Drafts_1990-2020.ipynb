```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import matplotlib.pyplot as plt
from unicodedata import normalize
import string
import html5lib
```

```python
lottery_range = list(np.arange(1990,2021))
```

```python
year = []
team_abb = []
player = []
college = []
draft_round = []
draft_pick = []

for i in lottery_range:
    url = f"https://www.basketball-reference.com/draft/NBA_{i}.html"
    response = requests.get(url)
    soup = bs(response.text, 'html.parser')
    left = soup.find_all('td', class_='left')
    pick_counter = 0
    for j in range(0, len(left), 3):
        team_abb.append(left[j].text)
        player.append(left[j+1].text)
        college.append(left[j+2].text)
        year.append(i)
        pick_counter += 1
        draft_pick.append(pick_counter)
        if pick_counter <= (len(left)/6):
            draft_round.append(1)
        else:
            draft_round.append(2)
    
    
```

```python
draft_df = pd.DataFrame({'year':year,'team_abb':team_abb,'player':player,'college':college,'round':draft_round,'pick':draft_pick})
```

```python
draft_df.to_csv('1990_2020drafts.csv',index=False)
```
