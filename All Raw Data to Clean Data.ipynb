```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import string
pd.set_option('display.max_columns', None)
```

```python
init_df = pd.read_csv('/Users/elliottlevy/sda/slam-dunk-analytics/Players_A-E.csv')
f_df = pd.read_csv('/Users/elliottlevy/sda/slam-dunk-analytics/Players_f.csv')
gl_df = pd.read_csv('/Users/elliottlevy/sda/slam-dunk-analytics/players_g_to_l')
ms_df = pd.read_csv('/Users/elliottlevy/sda/slam-dunk-analytics/Players_M-S.csv')
tz_df = pd.read_csv('/Users/elliottlevy/sda/slam-dunk-analytics/TtoZ.csv')
```

```python
total_df = pd.concat([init_df, f_df, gl_df, ms_df, tz_df])
total_df
```

```python
total_df = total_df.drop(['Unnamed: 0','Unnamed: 0.1'], axis=1)
total_df
```

```python
total_df = total_df.rename(columns = {"FG%":"FGPCT", "2P%":"2PPCT", "3P%":"3PPCT", "FT%":"FTPCT"})
total_df
```

```python
total_df = total_df.reset_index(drop=True)
total_df
```

```python
last_year = []

for i in range(len(total_df['Season'])):
    if total_df['Season'][i] == 'Career':
        last_year.append(total_df['Season'][i-1])
    else:
        pass
```

```python
career_df = total_df[total_df['Season'] == 'Career']
career_df
```

```python
career_df['last_season'] = last_year
```

```python
career_df = career_df.reset_index(drop=True)
career_df
```

```python
career_df['unique_player_key'] = career_df["player"] + ", " + career_df["School"] + ", " + career_df['last_season']
```

```python
last_year_update = []

for i in last_year:
    year_split = i.split('-')
    draft_year = int(year_split[0]) + 1
    last_year_update.append(str(draft_year))
    
```

```python
career_df['draft_year'] = last_year_update
```

```python
career_df['official_key'] = career_df["player"] + ", " + career_df["School"] + ", " + career_df['draft_year']
career_df
```

```python
career_df.to_csv('career.csv',index=False)
```
