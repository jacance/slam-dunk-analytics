```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
careersql = pd.read_csv('career.csv')
draftsql = pd.read_csv('1993later.csv')
```

```python
combined_df = pd.merge(careersql, draftsql, how='left', left_on='new_key', right_on='official_key')

combined_df
```

```python
combined_df.columns
```

```python
combined_df = combined_df[['School', 'Conf', 'G', 'GS', 'MP',
       'FG', 'FGA', 'FGPCT', '2P', '2PA', '2PPCT', '3P', '3PA', '3PPCT', 'FT',
       'FTA', 'FTPCT', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS',
       'player_x', 'last_season', 'unique_player_key', 'draft_year', 'new_key',
       'round', 'pick',]]
combined_df
```

```python
combined_df = combined_df.fillna(0)
combined_df
```

```python
combined_df['round'] = combined_df['round'].astype('int64')
combined_df['pick'] = combined_df['pick'].astype('int64')
```

```python
combined_df.columns
```

```python
from tensorflow.keras.models import Sequential

model2 = Sequential()
```

```python
from tensorflow.keras.layers import Dense
number_inputs = 22
number_hidden_nodes = 8
model2.add(Dense(units=number_hidden_nodes,
                activation='relu', input_dim=number_inputs))
model2.add(Dense(units=number_hidden_nodes, activation='relu'))
model2.add(Dense(units=number_hidden_nodes, activation='relu'))
model2.add(Dense(units=number_hidden_nodes, activation='relu'))
number_classes = 3
model2.add(Dense(units=number_classes, activation='softmax'))
```

```python
model2.summary()
```

```python
model2.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model2.fit(
    X_train_scaled,
    y_train,
    class_weight={0:1-(len(combined_df[combined_df['round']==0])/len(combined_df)), 1:1-(len(combined_df[combined_df['round']==1])/len(combined_df)), 2:1-(len(combined_df[combined_df['round']==2])/len(combined_df))},
    epochs=50,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model2.evaluate(
    X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
