## Neural Network

```python
from tensorflow.keras.models import Sequential

model2 = Sequential()
```

```python
from tensorflow.keras.layers import Dense
number_inputs = 22
number_hidden_nodes = 8
model2.add(Dense(units=number_hidden_nodes,
                activation='relu', input_dim=number_inputs))
model2.add(Dense(units=number_hidden_nodes, activation='relu'))
model2.add(Dense(units=number_hidden_nodes, activation='relu'))
model2.add(Dense(units=number_hidden_nodes, activation='relu'))
number_classes = 3
model2.add(Dense(units=number_classes, activation='softmax'))
```

```python
model2.summary()
```

```python
model2.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model2.fit(
    X_train_scaled,
    y_train,
    epochs=50,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model2.evaluate(
    X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
import numpy as np
train_predictions = model2.predict_classes(X_train_scaled)
test_predictions = model2.predict_classes(X_test_scaled)
print(f"Train: {train_predictions}")
print(f"Test: {test_predictions}")
```

## RFE

```python
%time

from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline

max_scores = []

for i in range(1, len(X.columns)):
    rfe = RFE(estimator=DecisionTreeClassifier(), n_features_to_select=i)
    model = DecisionTreeClassifier()
    pipeline = Pipeline(steps=[('s',rfe),('m',model)])
    # evaluate model
    cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=1)
    n_scores = cross_val_score(pipeline, Xnot2021_train_scaled, ynot2021_train, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
    # report performance
    max_scores.append(mean(n_scores))
    print(mean(n_scores), i)
```

```python
%time

import scipy.stats as ss

score_ranks = ss.rankdata(max_scores)
n_features = list(score_ranks).index(len(max_scores)-1) + 1
n_features
```
